```python
doctor_df = pd.DataFrame(columns=['name', 'int', 'area'])
doctor_df['name'] = doctor_name_list
doctor_df
```

```python
import random

area = ['piplod', 'vesu', 'pal', 'adajan', 'bhatar', 'ghod dod road', 'katargam', 'udhna', 'khajod', 'kamrej']

int_choice = ['A', 'B', 'C', 'D', 'E','F', 'G', 'H', 'I', 'J']

for i in range(len(doctor_df)):
    doctor_df['area'][i] = random.choice(area)
    doctor_df['int'][i] = random.choice(int_choice)
```

```python
import numpy as np
```

```python
# Doctors recommendation based on top ratings...
def recommend(int):
    # fetching the index of so and so specialization --> (int)
    index = np.where(doctor_df['int'].index.values == int)[0][0]
    similar_items = sorted(list(enumerate(new_similarity_scores[index])), key = lambda x : x[1], reverse=True)[1:10]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = doctor_df[doctor_df['name'] == new_pt.index[i[0]]]
        
        item.extend(list(temp_df['name'].values))
        item.extend(list(temp_df['int'].values))
        item.extend(list(temp_df['area'].values))
        data.append(item)
        
    return data
        
```

```python
recommend('Dr. Victoria Smith')
```

```python
# recommending users the doctors in their area based on their input symptons...

def recommend(int, area):
    doctor_df[['int', 'area']] == (int, area)
    groups = doctor_df.groupby(['int'])
    result = groups.get_group(int)
    return result
```
